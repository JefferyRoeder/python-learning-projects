```python
#3 Write a Pandas program to extract first 7 records from employees file.

employees = pd.read_csv('employees.csv')
employees.head(7)
```

```python
#4 Write a Pandas program to select distinct department id from employees file

pd.Series(employees.department_id.unique())
```

```python
#5 Write a Pandas program to display the first and last name, and department number for all employees whose last name is "McEwen"

mcewen_surname = employees['last_name'] == 'McEwen'
```

```python
employees[['first_name','last_name','department_id']][mcewen_surname]
```

```python
#6 Write a Pandas program to display the first, last name, salary and department number for those employees whose first name starts with the letter 'S'

first_name_s = employees['first_name'].str[0] == 'S'
```

```python
employees[['first_name','last_name','department_id']][first_name_s]
```

```python
# 7 Write a Pandas program to display the first, last name, salary and department number for those employees whose first name does not contain the letter 'M'
first_name_not_M = employees['first_name'].str.lower().str.contains('m') == False
employees[first_name_not_M]
```

```python
#8 Write a Pandas program to display the first name, last name, salary and department number in ascending order by department numbe

employees[['first_name','last_name','salary','department_id']].sort_values(by='department_id',ascending=False)
```
